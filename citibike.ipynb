```python
# Dependencies
import pandas as pd
```

```python
# List of file names to be combined 
files15 = ['data/201501.csv', 'data/201502.csv', 'data/201503.csv', 'data/201504.csv', 'data/201505.csv', 'data/201506.csv', 'data/201507.csv', 'data/201508.csv', 'data/201509.csv', 'data/201510.csv', 'data/201511.csv', 'data/201512.csv']

# Create an empty list to store each dataframe
df15_list = []

# Loop through each file in the list
for x in files15:
    # Read the file into a dataframe
    df15 = pd.read_csv(x)

    # Append the dataframe to the list
    df15_list.append(df15)

# Concatenate all dataframes in the list into one dataframe
df15 = pd.concat(df15_list)

# Show dataframe
df15
```

```python
# Get the number of null values in each column 
null_counts = df15.isna().sum()

# Show the number of null values
null_counts
```

```python
# Drop rows containing null values
df15.dropna(inplace=True)

# Check to make sure all nulls were dropped
null_counts = df15.isna().sum()
null_counts
```

```python
# List of file names to be combined 
files16 = ['data/201601.csv', 'data/201602.csv', 'data/201603.csv', 'data/201604.csv', 'data/201605.csv', 'data/201606.csv', 'data/201607.csv', 'data/201608.csv', 'data/201609.csv', 'data/201610.csv', 'data/201611.csv', 'data/201612.csv']

# Create an empty list to store each dataframe
df16_list = []

# list of file names to be renamed
rename_files = ['data/201610.csv', 'data/201611.csv', 'data/201612.csv']

# Loop through each file in the list
for y in files16:
    # Read the file into a dataframe
    df16 = pd.read_csv(y)

    # Check if the file name is in the list of files to be renamed
    if y in rename_files:
        # Specify the new column names
        new_columns = {'Trip Duration': 'tripduration','Start Time': 'starttime','Stop Time': 'stoptime','Start Station ID':'start station id','Start Station Name':'start station name','Start Station Latitude': 'start station latitude','Start Station Longitude': 'start station longitude','End Station ID': 'end station id','End Station Name': 'end station name','End Station Latitude': 'end station latitude','End Station Longitude': 'end station longitude','Bike ID': 'bikeid','User Type': 'usertype','Birth Year':'birth year','Gender': 'gender'}

        # Change the column names
        df16.rename(columns=new_columns, inplace=True)



    # Append the dataframe to the list
    df16_list.append(df16)

# Concatenate all dataframes in the list into one dataframe
df16 = pd.concat(df16_list)

# Show dataframe
df16
```

```python
# Get the number of null values in each column 
null_counts = df16.isna().sum()

# Show the number of null values
null_counts
```

```python
# Drop rows containing null values
df16.dropna(inplace=True)

# Check to make sure all nulls were dropped
null_counts = df16.isna().sum()
null_counts
```

```python
# Concatenate all dataframes for all the years into one
combined_df = pd.concat([df15, df16, df17, df18, df19, df20], axis=0)

# Display the results
combined_df
```
